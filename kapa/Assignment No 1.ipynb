## Data Normalization

**Scaling or Feature Scaling** is the process of changinng the scale of certain features to a common one. This is typically achieved through normalization and standardization.

**Normalization** is the process of scaling data into a range of [0, 1]. It's more useful and common for regression tasks.

x′ =x−xmin / xmax − xmin

**Standardization** is the process of scaling data so that they have a mean value of 0 and a standard deviation of 1. It's more useful and common for classification tasks.

x′=x−μ / σ
Where μ is mean and σ is standard deviation



```python
import pandas
import scipy
import numpy
from sklearn.preprocessing import MinMaxScaler

# data values
X = [ [110, 200], [120, 800], [310, 400], [140, 900], [510, 200], [653, 400] ,[310, 880] ]

# transofrm data
scaler = MinMaxScaler(feature_range=(0,5))
rescaledX = scaler.fit_transform(X)
```

```python
rescaledX
```

### **StandardScaler**

```python
from sklearn.preprocessing import StandardScaler
import pandas
import numpy

# data values
X = [ [110, 200], [120, 800], [310, 400], [140, 900], [510, 200], [653, 400] ,[310, 880] ]

# scaler
scaler = StandardScaler().fit(X)
rescaledX = scaler.transform(X)
```

### Normalize data

```python
from sklearn.preprocessing import Normalizer
import pandas
import numpy

# data values
X = [ [110, 200], [120, 800], [310, 400], [140, 900], [510, 200], [653, 400] ,[310, 880] ]

# normalize values
scaler = Normalizer().fit(X)
normalizedX = scaler.transform(X)
```

### Binary Data Transformation

```python
from sklearn.preprocessing import Binarizer
import pandas
import numpy

# data values
X = [ [501, 200], [120, 800], [310, 400], [140, 900], [510, 200], [653, 400] ,[310, 880] ]

# binarize data
binarizer = Binarizer(threshold=500).fit(X)
binaryX = binarizer.transform(X)
```

# 6.Turn categorical variables into quantitative variables in Python.

```python
import pandas as pd
import numpy as np
# Read in the CSV file and convert "?" to NaN
headers = ["symboling", "normalized_losses", "make", "fuel_type", "aspiration",
           "num_doors", "body_style", "drive_wheels", "engine_location",
           "wheel_base", "length", "width", "height", "curb_weight",
           "engine_type", "num_cylinders", "engine_size", "fuel_system",
           "bore", "stroke", "compression_ratio", "horsepower", "peak_rpm",
           "city_mpg", "highway_mpg", "price"]
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data",header=None, names=headers, na_values="?" )
df.head()
```

```python
df.dtypes
```

```python
obj_df = df.select_dtypes(include=['object']).copy()
obj_df.head()
```

```python
obj_df[obj_df.isnull().any(axis=1)]
```

```python
obj_df["num_doors"].value_counts()
```

```python
obj_df = obj_df.fillna({"num_doors": "four"})
```

```python
obj_df[obj_df.isnull().any(axis=1)]
```

### Approach #1 - Find and Replace
https://pbpython.com/categorical-encoding.html



```python
obj_df["num_cylinders"].value_counts()
```

```python
cleanup_nums = {"num_doors":     {"four": 4, "two": 2},
                "num_cylinders": {"four": 4, "six": 6, "five": 5, "eight": 8,
                                  "two": 2, "twelve": 12, "three":3 }}
```

```python
obj_df = obj_df.replace(cleanup_nums)
obj_df.head()
```

```python
obj_df.dtypes
```

### Approach #2 - Label Encoding

```python
obj_df["body_style"].value_counts()
```

```python
obj_df["body_style"] = obj_df["body_style"].astype('category')
obj_df.dtypes
```

```python
obj_df["body_style_cat"] = obj_df["body_style"].cat.codes
obj_df.head()
```

### Approach #3 - One Hot Encoding

```python
pd.get_dummies(obj_df, columns=["drive_wheels"]).head()
```

### Approach #4 - Scikit-Learn: OrdinalEncoder and OneHotEncoder

```python
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()
obj_df["make_code"] = ord_enc.fit_transform(obj_df[["make"]])
obj_df[["make", "make_code"]].head(11)
```

```python
from sklearn.preprocessing import OneHotEncoder

oe_style = OneHotEncoder()
oe_results = oe_style.fit_transform(obj_df[["body_style"]])
pd.DataFrame(oe_results.toarray(), columns=oe_style.categories_).head()
```

```python
obj_df = obj_df.join(pd.DataFrame(oe_results.toarray(), columns=oe_style.categories_))
```
